```python
import spacy
import requests
import bs4 as bs
import re
from spacy import displacy
```

```python
NLP = spacy.load("en_core_web_sm")
```

```python
Text1 = "Jakarta, Indonesia CNN Thirteen climbers have died and another 10 are missing in Indonesia following the eruption of a highly active volcano in the province of West Sumatra, officials said on Tuesday"
```

```python
#Preprocessing the text
Text1 = re.sub(r'^[,]', ' ', Text1)
Text1 = re.sub(r'\s+', ' ', Text1)

text_1 = NLP(Text1)
```

```python
for word in text_1.ents:
  print(word.text,word.label_)
```

```python
spacy.explain("DATE")
```

```python
spacy.explain("GPE")
```

```python
spacy.explain("NORP")
```

```python
spacy.explain("CARDINAL")
```

```python
spacy.explain("ORG")
```

```python
hasil_Text1 = displacy.render(text_1,style="ent",jupyter=True)
```

```python
import requests
import bs4 as bs
```

```python
source = "https://en.tempo.co/read/1806212/death-toll-from-indonesias-marapi-volcano-eruption-hits-23-rescue-mission-completed"
```

```python
content = requests.get(source).text

soup = bs.BeautifulSoup(content,'lxml')

body = soup.body.text
```

```python
import re
```

```python
Text2 = re.sub(r'^[,]', ' ', body)
Text2 = re.sub(r'\s+', ' ', Text2)
```

```python
Text_2 = NLP(Text2)
```

```python
for word2 in Text_2.ents:
  print(word2.text,word2.label_)
```

```python
hasil_text2 = displacy.render(Text_2,style="ent", jupyter=True)
```
